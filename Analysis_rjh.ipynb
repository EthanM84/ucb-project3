```python
import pandas as pd
```

```python
# Read in data file #1
pd.options.display.max_colwidth = 200
occ_salary = pd.read_excel("Resources/occupation_salary_2.xlsx")
occ_salary
```

```python
# Get the columns into a list
occ_salary_cols = occ_salary.columns.tolist()

# Create a new DataFrame
occ_sal_df = pd.DataFrame(occ_salary.values, columns=occ_salary_cols)
print(occ_sal_df)
```

```python
# Read in data file #2
pd.options.display.max_colwidth = 200
auto_dat = pd.read_csv("Resources/automation_data_by_state_2.csv", encoding='ISO-8859-1')
auto_dat
```
